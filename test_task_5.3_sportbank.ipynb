```python
import pandas as pd 
import numpy as np
```

```python
data = pd.read_excel('~D:\W\Sportbank\data.xlsm')
```

```python
data['sessions_between_payments'] = (data.Sessions / data.Payments).apply(np.ceil)
```

```python
task3 = data.groupby('sessions_betwen_payments', as_index=False).aggregate({'id':'count'})\
    .rename(columns = {'id':'users_count'})
```

```python
task3.to_csv(r'D:\W\Sportbank\task3.csv', index = False, header=True)
```
