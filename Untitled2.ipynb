```python
import pandas as pd
import yfinance as yf
import seaborn as sns
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('~D:\W\Genesis test(Keiki)\data.txt', sep=";")
```

```python
#exchange rates
currency_EUR_to_USD = yf.Ticker('EURUSD=X').info['previousClose']
currency_AUD_to_USD = yf.Ticker('AUDUSD=X').info['previousClose']
currency_USD_to_UAH = yf.Ticker('USDUAH=X').info['previousClose']

print("currency EUR to USD: ", currency_EUR_to_USD, 
      "\ncurrency AUD_to USD: ", currency_AUD_to_USD, 
      "\ncurrency_USD_to_UAH: ", currency_USD_to_UAH)
```

```python
#converting data into a single currency(USD)
data['price'].where(~(data.currency == 'EUR'), other=data.price * currency_EUR_to_USD, inplace=True)
data['price'].where(~(data.currency == 'AUD'), other=data.price * currency_AUD_to_USD, inplace=True)
data['price'].where(~(data.currency == 'UAH'), other=data.price / currency_USD_to_UAH, inplace=True)

data.price = round(data.price, 2)
data.currency = 'USD'
```

```python
data.head(10)
```

```python
#find LT metric(full monthes)
LT = data[["user_id", "date"]]
LT["date"] = (pd.to_datetime(LT['date']) - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s')

LT = LT.groupby(["user_id"], as_index=False).apply(lambda x: (x.date.max() - x.date.min()))
LT.columns = ['user_id', 'number_of_months']
LT.number_of_months = LT.number_of_months // (30*24*60*60)
```

```python
#df of the total sum of purchases by each user
price_data = data[["user_id", "price"]]
price_data = price_data.groupby(["user_id"], as_index=False).sum()
```

```python
#create new df for saving metrics to calculate LTV
metrics = pd.merge(LT,price_data,on='user_id',how='outer')
#calculate ARPU(revenue of the period / number of clients during this period)
metrics = metrics.groupby(['number_of_months'], as_index=False).apply(lambda x: x.price.sum() / x.user_id.count())
metrics.columns = ['LT', 'ARPU']
#calculate LTV(LT * ARPU)
metrics['LTV'] = metrics.LT * metrics.ARPU

metrics
```
